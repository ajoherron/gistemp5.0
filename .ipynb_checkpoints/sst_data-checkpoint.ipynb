```python
sst_file = 'SST_data_trimmed.nc'
ds = xr.open_dataset(sst_file)
```

```python
# Create list of all time values
time_list = ds.time.values

# Select the SST data for the desired time coordinate
index = -1
time_i = time_list[index]
sst_data = ds.sel(time=time_i).sst

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sst_data.plot()
plt.title(f'Sea Surface Temperature (SST) on {time_i}')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
baseline_start = '1951-01-01'
baseline_end = '1980-12-31'
baseline_ds = ds['sst'].sel(time=slice(baseline_start, baseline_end))
monthly_climatology = baseline_ds.groupby('time.month').mean(dim='time')
anomaly_ds = ds.groupby('time.month') - monthly_climatology
```

```python
# Create list of all time values
time_list = anomaly_ds.time.values

# Select the SST data for the desired time coordinate
index = 1
time_i = time_list[index]
sst_data = anomaly_ds.sel(time=time_i).sst

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sst_data.plot()
plt.title(f'Sea Surface Temperature (SST) on {time_i}')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
